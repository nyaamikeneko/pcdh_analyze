```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings

# --- ユーザー環境に合わせて設定 ---
try:
    import _init_notebook
    from src.config import PROCESSED_DIR, SAMPLING_RATE
except ImportError:
    print("警告: _init_notebook または src.config が見つかりません。")
    print("デモ用にダミーのパスとサンプリング周波数を設定します。")
    from pathlib import Path
    PROCESSED_DIR = Path('./processed_data') # ダミーパス
    SAMPLING_RATE = 1000.0
    PROCESSED_DIR.mkdir(exist_ok=True)
# ------------------------------------

# PROCESSED_DIR内の全parquetファイルを取得
all_files = list(PROCESSED_DIR.glob('*.parquet'))
print(f"対象ファイル数: {len(all_files)}")

# QCに必要な列のみを指定
# EEGデータ列 (PFC_filtered など) を除外し、メモリ使用量を削減
columns_to_load = [
    'Event_ID', 'Stimulus_Type', 'Time_s', 
    'Subject_ID', 'Genotype'
]

if not all_files:
    print(f"警告: {PROCESSED_DIR} に .parquet ファイルが見つかりません。")
    qc_df = pd.DataFrame() # 空のDF
    problematic_files = pd.DataFrame() # 空のDF
    passed_files_list = [] # 空のリスト
else:
    qc_results = []
    
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)

        for data_path in all_files:
            print(f"QCチェック中: {data_path.name}") # どのファイルで失敗するか追跡
            try:
                # === メモリエラー対策： 必要な列だけを読み込む ===
                df = pd.read_parquet(data_path, columns=columns_to_load)
                # ===============================================

                # 基本情報
                subject_id = df['Subject_ID'].iloc[0]
                genotype = df['Genotype'].iloc[0]

                # --- 試行開始点の特定 ---
                event_starts_mask = (df['Event_ID'] > 0) & (df['Event_ID'].diff().fillna(1) != 0)
                trial_starts_df = df[event_starts_mask].copy()
                total_events = len(trial_starts_df)

                # 刺激タイプごとのカウント
                stim_counts = trial_starts_df['Stimulus_Type'].value_counts()

                # --- ISI（試行間隔）の計算 ---
                isi = trial_starts_df['Time_s'].diff().dropna()

                # --- 刺激長（Event_ID > 0 が続く時間）の確認 ---
                event_blocks = (df['Event_ID'].diff() != 0).cumsum()
                stim_durations_series = df[df['Event_ID'] > 0].groupby(event_blocks)['Time_s'].apply(
                    lambda x: x.max() - x.min() + (1.0 / SAMPLING_RATE)
                )

                # --- 600回ごとのインターバル確認 ---
                interval_times = {}
                try:
                    t_600_end = df[df['Event_ID'] == 600]['Time_s'].max()
                    t_601_start = df[df['Event_ID'] == 601]['Time_s'].min()
                    interval_times['Interval_600'] = t_601_start - t_600_end
                except (ValueError, TypeError):
                    interval_times['Interval_600'] = np.nan

                try:
                    t_1200_end = df[df['Event_ID'] == 1200]['Time_s'].max()
                    t_1201_start = df[df['Event_ID'] == 1201]['Time_s'].min()
                    interval_times['Interval_1200'] = t_1201_start - t_1200_end
                except (ValueError, TypeError):
                    interval_times['Interval_1200'] = np.nan

                qc_results.append({
                    'filename': data_path.name,
                    'Subject_ID': subject_id,
                    'Genotype': genotype,
                    'Total_Trials': total_events,
                    'Light_Trials': stim_counts.get('Light', 0),
                    'Sound_Trials': stim_counts.get('Sound', 0),
                    'Light+Sound_Trials': stim_counts.get('Light+Sound', 0),
                    'ISI_mean (s)': isi.mean() if not isi.empty else np.nan,
                    'ISI_median (s)': isi.median() if not isi.empty else np.nan,
                    'ISI_std (s)': isi.std() if not isi.empty else np.nan,
                    'Stim_Duration_mean (s)': stim_durations_series.mean() if not stim_durations_series.empty else np.nan,
                    'Stim_Duration_median (s)': stim_durations_series.median() if not stim_durations_series.empty else np.nan,
                    'Interval_600 (s)': interval_times['Interval_600'],
                    'Interval_1200 (s)': interval_times['Interval_1200'],
                })

            except Exception as e:
                print(f"  エラー: {data_path.name} の処理中に致命的な問題が発生しました: {e}")
                qc_results.append({
                    'filename': data_path.name,
                    'Subject_ID': 'Error',
                    'Genotype': 'Error',
                    'Total_Trials': 'Error',
                })
            
            # (オプション) メモリを明示的に解放
            del df
            import gc
            gc.collect()

    # 結果をDataFrameに変換
    qc_df = pd.DataFrame(qc_results)
    print("\n--- QCサマリーテーブル ---")
    display(qc_df)

    # --- QC結果の可視化 ---
    # (可視化コードは変更なし)
    # ... (前回のコードと同じ) ...
    # 総試行回数
    plt.figure(figsize=(10, 5))
    sns.histplot(data=qc_df, x='Total_Trials', hue='Genotype', multiple='stack', bins=30)
    plt.title('Distribution of Total Trials per Subject')
    plt.axvline(1800, color='r', linestyle='--', label='Expected 1800')
    plt.legend()
    plt.show()

    # ISI（試行間隔）
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=qc_df, x='Genotype', y='ISI_median (s)')
    plt.title('Inter-Stimulus Interval (ISI) Median')
    plt.axhline(1.0, color='r', linestyle='--', label='Expected 1.0s')
    plt.legend()
    plt.show()

    # 刺激長
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=qc_df, x='Genotype', y='Stim_Duration_median (s)')
    plt.title('Stimulus Duration Median')
    plt.axhline(0.5, color='r', linestyle='--', label='Expected 0.5s')
    plt.legend()
    plt.show()

    # --- 問題のあるファイルの特定 ---
    expected_trials = 1800
    trial_tolerance = 0.05
    isi_tolerance = 0.1
    dur_tolerance = 0.05

    qc_df_numeric = qc_df.apply(pd.to_numeric, errors='coerce').fillna(qc_df)

    is_error = qc_df_numeric['Total_Trials'] == 'Error'
    bad_trials = (qc_df_numeric['Total_Trials'] < expected_trials * (1 - trial_tolerance)) | \
                 (qc_df_numeric['Total_Trials'] > expected_trials * (1 + trial_tolerance))
    bad_isi = (qc_df_numeric['ISI_median (s)'] < 1.0 - isi_tolerance) | \
              (qc_df_numeric['ISI_median (s)'] > 1.0 + isi_tolerance)
    bad_duration = (qc_df_numeric['Stim_Duration_median (s)'] < 0.5 - dur_tolerance) | \
                   (qc_df_numeric['Stim_Duration_median (s)'] > 0.5 + dur_tolerance)

    problematic_files = qc_df[is_error | bad_trials | bad_isi | bad_duration]

    print("\n--- 問題の可能性があるファイル（分析から除外推奨） ---")
    display(problematic_files)

    # QCをパスしたファイルのリストを作成
    passed_files_df = qc_df.drop(problematic_files.index)
    passed_files_list = [PROCESSED_DIR / fname for fname in passed_files_df['filename']]
    print(f"\nQCをパスしたファイル数: {len(passed_files_list)}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pathlib import Path

# --- 環境設定（前回のコードから流用） ---
try:
    # _init_notebook をインポート（ユーザー環境に依存）
    import _init_notebook
    from src.config import PROCESSED_DIR, SAMPLING_RATE
except ImportError:
    print("警告: _init_notebook または src.config が見つかりません。")
# ------------------------------------


# --- 検証したいファイル名を指定 ---
# (例: 'wt262avs.adicht_rec2.parquet' など、meanとmedianが乖離したファイル)
filename_to_check = 'het165avstake2_rest.adicht_rec2.parquet' # ここを対象のファイル名に変更してください

data_path_to_check = PROCESSED_DIR / filename_to_check

if not data_path_to_check.exists():
    print(f"エラー: 指定されたファイルが見つかりません: {data_path_to_check}")
    print("デモを続行できません。filename_to_check を正しいパスに変更してください。")
else:
    print(f"--- 詳細検証: {filename_to_check} ---")

    try:
        # --- 修正点 ---
        # メモリ節約のため、検証に必要なカラムのみを読み込む
        columns_to_read = ['Event_ID', 'Time_s']
        print(f"メモリ節約のため、カラム {columns_to_read} のみを読み込みます...")
        df_check = pd.read_parquet(data_path_to_check, columns=columns_to_read)
        print("ファイル読み込み完了。")
        # ----------------

        # --- 1. ISI (SOA) の全データ取得 ---
        print("ISI (SOA) を計算中...")
        event_starts_mask_check = (df_check['Event_ID'] > 0) & (df_check['Event_ID'].diff().fillna(1) != 0)
        trial_starts_df_check = df_check[event_starts_mask_check].copy()
        
        # インターバル（600回ごと）も含むISI (SOA)
        isi_soa_series = trial_starts_df_check['Time_s'].diff().dropna()
        
        # インターバルを除外したISI (SOA)
        interval_indices = trial_starts_df_check[
            (trial_starts_df_check['Event_ID'] == 601) | 
            (trial_starts_df_check['Event_ID'] == 1201)
        ].index
        
        isi_soa_series_cleaned = trial_starts_df_check.drop(index=interval_indices)['Time_s'].diff().dropna()

        # --- 2. Duration の全データ取得 ---
        print("Duration を計算中...")
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=RuntimeWarning)
            
            event_blocks_check = (df_check['Event_ID'].diff() != 0).cumsum()
            duration_series = df_check[df_check['Event_ID'] > 0].groupby(event_blocks_check)['Time_s'].apply(
                lambda x: x.max() - x.min() + (1.0 / SAMPLING_RATE) if not x.empty else np.nan
            ).dropna()

        # --- 3. 可視化と比較 ---
        print("結果の表示...")
        
        print("\n--- ISI (SOA) 詳細 ---")
        print("【インターバル混入時】(meanがずれる原因):")
        display(isi_soa_series.describe())
        print("\n【インターバル除外時（試行間のみ）】(こちらが本来のISI/SOA):")
        display(isi_soa_series_cleaned.describe())

        print("\n--- Duration 詳細 ---")
        display(duration_series.describe())

        fig, axes = plt.subplots(1, 2, figsize=(15, 6))

        # ISI (SOA) のヒストグラム
        sns.histplot(isi_soa_series, bins=100, ax=axes[0], color='blue', label='Intervals Included')
        sns.histplot(isi_soa_series_cleaned, bins=100, ax=axes[0], color='orange', label='Intervals Excluded')
        axes[0].set_title(f'ISI (SOA) Distribution\nExpected Median = 1.0s (仮定)')
        axes[0].set_xlabel('Time (s)')
        # 5秒以上（インターバルと想定）は除外して表示
        axes[0].set_xlim(isi_soa_series_cleaned.min() * 0.9, isi_soa_series_cleaned.max() * 1.1) 
        axes[0].legend()

        # Duration のヒストグラム
        sns.histplot(duration_series, bins=50, ax=axes[1], color='green')
        axes[1].set_title(f'Stimulus Duration Distribution\nExpected = 0.5s')
        axes[1].set_xlabel('Time (s)')
        # 0.5秒前後を拡大
        axes[1].set_xlim(duration_series.min() * 0.99, duration_series.max() * 1.01)

        plt.suptitle(f'File: {filename_to_check}')
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"エラー: {e}")
```
